##4) Shuffle and Split training data to train (80%) and validation (20%) 

[5pts] Shuffle data

[5pts] split data

*Note: Solutions can vary

```
from sklearn.utils import shuffle

# Shuffle the training data
from sklearn.utils import shuffle
data_train = shuffle(data_train)

# Split 
splt = int(len(data_train)*.8)
split_data_train = data_train[1:splt]
split_data_valid = data_train[splt+1:]
```

##5) Train DecisionTreeClassifier 

[5pts] split train data to attributes and label

[5pts] split validation data to attribute and label *can be added in the next section*

[5pts] Train tree 1 (using default or specify min_samples_split=1.0) both are correct

[5pts] Train tree 2

[5pts] Train tree 3

```
# To train Decision Tree you need two variables (1) attributes (2) labels

# Split your training data to "attributes" and "labels" 

tr_lbl  = split_data_train['label']
tr_data = split_data_train.drop('label', axis=1)
val_lbl  = split_data_valid['label']
val_data = split_data_valid.drop('label', axis=1)



# Train Decision tree classifier
from sklearn.tree import DecisionTreeClassifier


dtree1 = DecisionTreeClassifier()
dtree1.fit(tr_data, tr_lbl)

dtree2 = DecisionTreeClassifier(min_samples_split=3)
dtree2.fit(tr_data, tr_lbl)

dtree3 = DecisionTreeClassifier(min_samples_split=5)
dtree3.fit(tr_data, tr_lbl)
```

## 6) Testing your trained classifier on Validation set

[6pts] predicting validation labels on tree 1 and calculate accuracy using predicted labels and true labels


[6pts] predicting validation labels on tree 2 & calculate accuracy using predicted labels and true labels


[6pts] predicting validation labels on tree 3 & calculate accuracy using predicted labels and true labels



```
from sklearn.metrics import accuracy_score


pred_val = dtree1.predict(val_data)
print(accuracy_score(val_lbl, pred_val)*100)

pred_val = dtree2.predict(val_data)
print(accuracy_score(val_lbl, pred_val)*100)

pred_val = dtree3.predict(val_data)
print(accuracy_score(val_lbl, pred_val)*100)
```

## 7) Testing your trained classifier on Test set

[10pts] Reading test data

```
# Load test data
data_test = pd.read_csv("drive/My Drive/ML412_Fall2018_2019/HW1/credit-german-test_num.csv") 


# Predict
te_lbl  = data_test['label']
te_data = data_test.drop('label', axis=1)

pred_te = dtree1.predict(te_data)
accuracy_score(te_lbl, pred_te)*100
```
